```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/instacart/interaction.csv', sep='\t')
test_df = df[df['x_label'] == 1]
test_df
```

```python
test_df = test_df.sort_values('order_id')
test_df
```

```python
test_df = test_df[['order_id', 'aisle_id', 'x_label']]
test_df
```

```python
test_df['aisle_id'].max()
```

```python
cates = test_df['aisle_id'].to_numpy()
orders = test_df['order_id'].to_numpy()
print(
    orders,
    cates,
)
```

```python
l = len(test_df)
cate_dict = {}

for i in range(0, l):
    cate_dict[orders[i]] = []

for i in range(0, l):
    cate_dict[orders[i]].append(cates[i])
```

```python
cate_dict
```

```python
test_data = []

for i in cate_dict.values():
    test_data.append(i)

test_data
```

```python
np.save('data/instacart/test_data.npy', test_data)
```
