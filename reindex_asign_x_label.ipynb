```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/instacart/train_user2.csv', sep=',')
test = pd.read_csv('data/instacart/test_user2.csv', sep=',')
```

```python
temp = train.sort_values('user_id')
temp.head(20)
```

```python
print(train['aisle_id'].max(),
      test['aisle_id'].max())
```

```python
train['x_label'] = [0] * len(train)
test['x_label'] = [1] * len(test)
```

```python
df = pd.concat([train, test]).sort_values('user_id')
df.head()
```

```python
count_u_id = df.groupby('user_id').count()
count_u_id = count_u_id['aisle_id'].to_numpy()
count_u_id
```

```python
len(count_u_id)
```

```python
start_id = 0
new_u_id = []
for i in range(0, len(count_u_id)):
    new_u_id.extend([start_id] * count_u_id[start_id])
    start_id+=1

df['user_id'] = new_u_id
df['aisle_id'] -= 1
```

```python
df
```

```python
df = df[['user_id', 'aisle_id']].drop_duplicates()
```

```python
df.to_csv('data/instacart/interaction.csv', sep='\t', index=False)
```
